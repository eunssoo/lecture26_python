리스트 만들기

```python
a = []
b = list()
c = [1, 2, 3]
d = ['banana', 'apple', 'tomato', 'watermelon']
e = [1, 2, 3, 'one', 'two', 'three']
f = [1, 2, ['banana', 100], ['tomato', 300]]

print('a =', a)
print('b =', b)
print('c =', c)
print('d =', d)
print('e =', e)
print('f =', f)
```

리스트의 항목에 접근하기 인덱싱

```python
test_list = [1, 2, 3, 'one', 'two', 'three']

print(test_list[0])    # 1
print(test_list[3])    # one
print(test_list[-1])   # three
print(test_list[-3])   # one

for item in test_list:
    print(item)
```

리스트의 항목에 접근하기 슬라이싱

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[:5])
print(test_list[5:])
print(test_list[1:3])

print(test_list[::2])
print(test_list[::3])

print(test_list[:])
print(test_list[::])

print(test_list[-1:-7:-2])
print(test_list[1:-1])
print(test_list[-3::-1])
print(test_list[::-1])
print(test_list[::-2])
```

리스트에 항목 추가하기

```python
a_list = [1, 2, 3, 4]
a_list.append('four')
print(a_list)

a_list.insert(3, 'three')
print(a_list)

a_list.extend([10, 11])
print(a_list)

b_list = [1, 2, 3]
c_list = ['a', 'b', 'c']
d_list = b_list + c_list
print(d_list)

e_list = b_list * 3
print(e_list)
```

리스트에서 항목 삭제하기

```python
test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
item = test_list.pop(2)
print('pop(2)으로 삭제한 값:', item)
print('현재 리스트:', test_list)

test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
item = test_list.pop()
print('pop()으로 삭제한 값:', item)
print('현재 리스트:', test_list)

test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
test_list.remove('a')
print("remove('a') 후:", test_list)

test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
del test_list[2]
print('del test_list[2] 후:', test_list)

test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
del test_list[:2]
print('del test_list[:2] 후:', test_list)

test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]
test_list.clear()
print('clear() 후:', test_list)
```

리스트에서 항목 재배열하기

```python
a_list = ['one', 'two', 'three', 'four']
a_list.reverse()
print(a_list)

a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list, reverse=True)
print(a_list)
print(b_list)

a_list.sort(key=len)
print(a_list)
```

리스트 탐색하기

```python
test_list = [1, 2, 3, 1, 2, 3]

print(3 in test_list)
print(test_list.index(3))
print(test_list.count(3))

print(4 in test_list)
print(test_list.count(4))

print(len(test_list))
print(max(test_list))
print(min(test_list))
print(sum(test_list))
```

리스트 내포

```python
print([x for x in range(1, 6)])
print([x for x in range(1, 11) if x % 2 == 0])
print([x * x for x in range(1, 6)])
print([x for x in range(1, 101) if x % 7 == 0])
```
