```python
def gugudan(dan):
    print(f'[ 구구단 {dan}단 ]')
    for i in range(1, 10):
        print(f'{dan} x {i} = {dan * i}')
```

```python
# 2단부터 9단까지 출력
for num in range(2, 10):
    gugudan(num)
    print()
```

```python
# return 값 확인
def add_and_mnl(a, b):
    return a + b, a * b

result = add_and_mnl(3, 4)
print(result)

result1, result2 = add_and_mnl(3, 4)
print(result1, result2)
```

```python
# 지역변수, 전역변수
sum = 10
def test(sum, num):
    sum += num
    return sum
    
sum = test(sum, 5)
print(sum)
```

```python
# 지역변수, 전역변수
sum = 10
def test(num):
    global sum
    sum += num
    
test(5)
print(sum)
```
