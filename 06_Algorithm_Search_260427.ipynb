```python
#recursive call을 활용한 binary search
def binary_search_recursive(A, key, low, high):
    # binary_search 함수 정의
    # A    : 탐색할 정렬된 리스트
    # key  : 찾고 싶은 값
    # low  : 현재 탐색 범위의 시작 인덱스
    # high : 현재 탐색 범위의 끝 인덱스

    if low > high:
        # 종료 조건
        # low가 high보다 커졌다는 것은 더 이상 탐색할 범위가 없다는 뜻
        # 즉, key를 찾지 못했다는 의미
        return -1
    
    middle = (low + high) // 2
    # 현재 탐색 범위의 중앙 인덱스를 계산
    # //는 정수 나눗셈
    # 예: (0 + 6) // 2 = 3
    if key == A[middle]:
        # 찾고자 하는 key가 중앙값 A[middle]과 같다면
        # 탐색 성공
        return middle
        # key가 들어 있는 인덱스 middle을 반환

    elif key < A[middle]:
        # key가 중앙값보다 작다면
        # 정렬된 배열이므로 key는 middle의 왼쪽에만 있을 수 있음
        # 오른쪽 범위는 볼 필요가 없음

        return binary_search_recursive(A, key, low, middle - 1)
        # 탐색 범위를 low부터 middle - 1까지로 줄여서 다시 이진 탐색
        # 즉, 왼쪽 절반만 다시 탐색

    else:
        # 위 조건들이 아니라면 key > A[middle]인 경우
        # key가 중앙값보다 크다는 뜻
        # 정렬된 배열이므로 key는 middle의 오른쪽에만 있을 수 있음

        return binary_search_recursive(A, key, middle + 1, high)
        # 탐색 범위를 middle + 1부터 high까지로 줄여서 다시 이진 탐색
        # 즉, 오른쪽 절반만 다시 탐색

A = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]

key = int(input("찾을 값을 입력하세요: "))

result = binary_search_recursive(A, key, 0, len(A) - 1)

if result == -1:
    print("찾는 값이 없습니다.")
else:
    print(f"{key}는 인덱스 {result}에 있습니다.")
```

```python
# while loop를 활용한 binary search
def binary_search(A, key):
    # 이진 탐색 함수 정의
    # A   : 정렬된 리스트
    # key : 찾고 싶은 값

    low = 0
    # 탐색 범위의 시작 인덱스 (맨 왼쪽)

    high = len(A) - 1
    # 탐색 범위의 끝 인덱스 (맨 오른쪽)
    # 리스트는 0부터 시작하므로 길이 - 1

    while low <= high:
        # 탐색 범위가 존재하는 동안 반복
        # low가 high보다 커지면 탐색 종료

        middle = (low + high) // 2
        # 현재 탐색 범위의 중앙 인덱스 계산
        # // 는 정수 나눗셈 (소수점 버림)

        if A[middle] == key:
            # 중앙값이 찾고자 하는 값과 같다면
            return middle
            # 해당 인덱스를 반환하고 종료

        elif A[middle] > key:
            # 중앙값이 key보다 크다면
            high = middle - 1
            # key는 왼쪽에 있으므로
            # 탐색 범위를 왼쪽 절반으로 줄임

        else:
            # 중앙값이 key보다 작다면
            low = middle + 1
            # key는 오른쪽에 있으므로
            # 탐색 범위를 오른쪽 절반으로 줄임

    return -1
    # 반복문이 끝났다는 것은 값을 찾지 못했다는 의미
    # -1을 반환 (없음 표시)
    
# 데이터
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]

# 입력
num = int(input('찾을 숫자: '))

# 탐색 실행
result = binary_search(data, num)

# 출력
if result == -1:
    print("찾는 값이 없습니다.")
else:
    print(f"{num}는 인덱스 {result}에 있습니다.")
```

```python
def binary_search(A, key):
    low = 0
    high = len(A) - 1

    while low <= high:
        middle = (low + high) // 2

        if A[middle] == key:
            return middle
        elif A[middle] > key:
            high = middle - 1
        else:
            low = middle + 1

    return -1

test_data = [34, 2, 47, 15, 48, 0]
answer = [10, 0, 15, -1, -1, -1]

for i in range(6):
    if binary_search(data, test_data[i]) == answer[i]:
        print('pass : ', test_data[i])
    else:
        print('Fail : ', test_data[i])
```
