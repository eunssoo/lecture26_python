```python
print('--- 1. 튜플 만들기 ---')
t1 = (10, 20, 30)
t2 = ('apple', 'banana', 'cherry')
t3 = ()              # 빈 튜플
t4 = tuple()         # 빈 튜플
print(t1)
print(t2)
print(t3)
print(t4)
```

```python
print('\n--- 2. 원소가 1개인 튜플 ---')
a = (3)      # 그냥 int
b = (3,)     # 튜플
print(a, type(a))
print(b, type(b))
```

```python
print('\n--- 3. 괄호 없는 튜플 (packing) ---')
t5 = 1, 2, 3
print(t5, type(t5))
```

```python
print('\n--- 4. 튜플 언패킹 ---')
x, y, z = (100, 200, 300)
print(x)
print(y)
print(z)
```

```python
print('\n--- 5. 값 교환 (swap) ---')
a = 10
b = 20
print('교환 전:', a, b)
a, b = b, a
print('교환 후:', a, b)
```

```python
print('\n--- 6. 인덱싱 ---')
t = (10, 20, 30, 40, 50)
print(t[0])     # 첫 번째
print(t[2])     # 세 번째
print(t[-1])    # 마지막
print(t[-2])    # 뒤에서 두 번째
```

```python
print('\n--- 7. 슬라이싱 ---')
print(t[1:4])   # (20, 30, 40)
print(t[:3])    # (10, 20, 30)
print(t[2:])    # (30, 40, 50)
print(t[::2])   # (10, 30, 50)
print(t[::-1])  # 뒤집기
```

```python
print('\n--- 8. for문으로 접근 ---')
for item in t:
    print(item)
```

```python
print('\n--- 9. in 연산자 ---')
print(30 in t)   # True
print(99 in t)   # False
```

```python
print('\n--- 10. index(), count() ---')
t6 = (10, 20, 20, 30, 20)
print(t6.index(30))   # 3
print(t6.count(20))   # 3
```

```python
print('\n--- 11. len(), max(), min(), sum() ---')
nums = (5, 8, 2, 10, 3)
print('길이:', len(nums))
print('최댓값:', max(nums))
print('최솟값:', min(nums))
print('합계:', sum(nums))
```

```python
print('\n--- 12. sorted() ---')
t7 = (4, 1, 3, 2)
result = sorted(t7)
print('원본 튜플:', t7)
print('sorted 결과:', result)
print('sorted 결과의 타입:', type(result))   # list
```

```python
print('\n--- 13. + 연산 ---')
t8 = (1, 2, 3)
t9 = (4, 5, 6)
t10 = t8 + t9
print(t10)
```

```python
print('\n--- 14. * 연산 ---')
t11 = ('A', 'B')
t12 = t11 * 3
print(t12)
```

```python
print('\n--- 15. tuple()로 리스트를 튜플로 변환 ---')
lst = [1, 2, 3]
t13 = tuple(lst)
print(lst)
print(t13)
```

```python
print('\n--- 16. 함수에서 여러 값 반환 ---')
def calc(a, b):
    return a + b, a * b

result = calc(3, 4)
print('result:', result)
print('result 타입:', type(result))

sum_value, mul_value = calc(3, 4)
print('합:', sum_value)
print('곱:', mul_value)
```

```python
print('\n--- 17. 튜플은 수정 불가 ---')
t14 = (10, 20, 30)
print('원본:', t14)

# 아래 코드는 오류가 나므로 주석 처리
# t14[1] = 99          # TypeError
# t14.append(40)       # AttributeError
# t14.remove(20)       # AttributeError
# t14.sort()           # AttributeError

print('튜플은 항목 수정/추가/삭제가 불가능함')
```

```python
print('\n--- 18. del로 튜플 전체 삭제 ---')
t15 = (1, 2, 3)
print('삭제 전:', t15)
del t15
# print(t15)   # 삭제 후 출력하면 오류

print('t15는 del로 전체 삭제됨')
```
