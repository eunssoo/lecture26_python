```python
s = input()

substrings = set()

for i in range(len(s)):
    for j in range(i + 1, len(s) + 1):
        substrings.add(s[i:j])

print(len(substrings))
```

```python
s = input()

substrings = set()

for i in range(len(s)):
    for j in range(i, len(s)):
        sub_str = (s[i:j+1])
        substrings.add(sub_str)

print(len(substrings))
```

```python
n = int(input())

company = set()

for _ in range(n):
    name, status = input().split()

    if status == "enter":
        company.add(name)
    else:
        company.remove(name)

for name in sorted(company, reverse=True):
    print(name)
```
